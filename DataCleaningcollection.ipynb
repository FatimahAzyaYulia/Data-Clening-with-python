Imputasi

```python
# Melakukan analisa deskriptif
df_data.describe()
```

```python
# check data missing value
df_data.isnull().any()
```

```python
# Mengisi missing value pada kolom umur menggunakan nmedian
df_data['umur'] = df_data['umur'].fillna(df_data['umur'].median())
```

```python
# Mengisi missing value pada kolom jenis kelamin
df_data['jenis_kelamin'] = df_data['jenis_kelamin'].fillna('Unknown')
```

```python
# Mengisi missing value pada kolom pekerjaan
df_data['pekerjaan'] = df_data['pekerjaan'].fillna('Unknown')
```

```python
# Mengisi missing value pada kolom gaji menggunakan median
df_data['gaji'] = df_data['gaji'].fillna(df_data['gaji'].median())
```

```python
# Mengisi missing value pada kolom is menikah menggunakan nilai yg paling sering muncul (mode)
df_data['is_menikah'] = df_data['is_menikah'].fillna(df_data['is_menikah'].mode()[0])
```

```python
# Mengisi missing value pada kolom berat menggunakan median
df_data['berat'] = df_data['berat'].fillna(df_data['berat'].median())
```

```python
# Mengisi missing value pada kolom tinggi menggunakan mean
df_data['tinggi'] = df_data['tinggi'].fillna(df_data['tinggi'].mean())
```

```python
# Mengisi missing value pada kolom is merokok menggunakan nilai yg paling sering muncul (mode)
df_data['is_merokok'] = df_data['is_merokok'].fillna(df_data['is_merokok'].mode()[0])
```

```python
# Mengisi missing value pada kolom pendidikan
df_data['pendidikan'] = df_data['pendidikan'].fillna('Unknown')
```

```python
# check hasil missing value setelah melakukan imputasi
df_data.isnull().sum()
```

```python
# Melakukan analisa deskriptif untuk mendapatkan nilai q1 dan q3
df_data.describe()
```

```python
# Menghitung Kuartil 1 (Q1) dan Kuartil 3 (Q3) pada kolom umur
q1_umur = df_data['umur'].quantile(0.25)
q3_umur = df_data['umur'].quantile(0.75)
iqr_umur = q3_umur - q1_umur
print('Q1 Umur:', q1_umur)
print('Q3 Umur:', q3_umur)
print('IQR Umur:', iqr_umur)
```

```python
# Menghitung Kuartil 1 (Q1) dan Kuartil 3 (Q3) pada kolom Gaji
q1_gaji = df_data['gaji'].quantile(0.25)
q3_gaji = df_data['gaji'].quantile(0.75)
iqr_gaji = q3_gaji - q1_gaji
print('Q1 Gaji:', q1_gaji)
print('Q3 Gaji:', q3_gaji)
print('IQR Gaji:', iqr_gaji)
```

```python
# Menghitung Kuartil 1 (Q1) dan Kuartil 3 (Q3) pada kolom berat
q1_berat = df_data['berat'].quantile(0.25)
q3_berat = df_data['berat'].quantile(0.75)
iqr_berat = q3_berat - q1_berat
print('Q1 Berat:', q1_berat)
print('Q3 Berat:', q3_berat)
print('IQR Berat:', iqr_berat)
```

```python
# Menghitung Kuartil 1 (Q1) dan Kuartil 3 (Q3) pada kolom tinggi
q1_tinggi = df_data['tinggi'].quantile(0.25)
q3_tinggi = df_data['tinggi'].quantile(0.75)
iqr_tinggi = q3_tinggi - q1_tinggi
print('Q1 Tinggi:', q1_tinggi)
print('Q3 Tinggi:', q3_tinggi)
print('IQR Tinggi:', iqr_tinggi)
```

```python
# Find lower and upper bound
lower_bound_umur = q1_umur - 1.5 * iqr_umur
upper_bound_umur = q3_umur + 1.5 * iqr_umur

lower_bound_gaji = q1_gaji - 1.5 * iqr_gaji
upper_bound_gaji = q3_gaji + 1.5 * iqr_gaji

lower_bound_berat = q1_berat - 1.5 * iqr_berat
upper_bound_berat = q3_berat + 1.5 * iqr_berat

lower_bound_tinggi = q1_tinggi - 1.5 * iqr_tinggi
upper_bound_tinggi = q3_tinggi + 1.5 * iqr_tinggi

print('Lower Bound Umur:', lower_bound_umur)
print('Upper Bound Umur:', upper_bound_umur)
print('Lower Bound Gaji:', lower_bound_gaji)
print('Upper Bound Gaji:', upper_bound_gaji)
print('Lower Bound Berat:', lower_bound_berat)
print('Upper Bound Berat:', upper_bound_berat)
print('Lower Bound Tinggi:', lower_bound_tinggi)
print('Upper Bound Tinggi:', upper_bound_tinggi)
```

```python
# Mengambil outlier yg dibawah lower bound
df_data[df_data['umur'] < lower_bound_umur]
df_data[df_data['gaji'] < lower_bound_gaji]
df_data[df_data['berat'] < lower_bound_berat]
df_data[df_data['tinggi'] < lower_bound_tinggi]
```

```python
# Mengambil outlier yang di atas upper bound
df_data[df_data['umur'] > upper_bound_umur]
df_data[df_data['gaji'] > upper_bound_gaji]
df_data[df_data['berat'] > upper_bound_berat]
df_data[df_data['tinggi'] > upper_bound_tinggi]
```

```python
# Memfilter data dari outlier
df_clean = df_data[(df_data['umur'] >= lower_bound_umur) & (df_data['umur'] <= upper_bound_umur)]
df_clean  = df_data[(df_data['gaji'] >= lower_bound_gaji) & (df_data['gaji'] <= upper_bound_gaji)]
df_clean  = df_data[(df_data['berat'] >= lower_bound_berat) & (df_data['berat'] <= upper_bound_berat)]
df_clean  = df_data[(df_data['tinggi'] >= lower_bound_tinggi) & (df_data['tinggi'] <= upper_bound_tinggi)]
df_clean
```
